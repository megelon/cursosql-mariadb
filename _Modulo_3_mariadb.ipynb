Cada tabla consta de muchas filas y columnas .

MariaDB se creó originalmente como un reemplazo directo de MySQL, por lo que MariaDB y MySQL están particularmente cerca.

- El modelo relacional permite que cualquier archivo se relacione con cualquier otro por medio de un campo común. 
- De repente, la complejidad del diseño se redujo versus los otros modelos. 
- Que debían confiar en relaciones padre-hijo o propietario-miembro.

- EL acceso no es a través de rutas hacia y desde los archivos, sino desde una relación directa entre archivos, por ello se pueden agregar fácilmente nuevas relaciones entre estos archivos.

- De hecho, en 1970, cuando EF Codd desarrolló el modelo relacional, se pensó que no era práctico. 
- La mayor facilidad de uso conlleva una penalización de rendimiento significativa.
- Y el hardware de esos días no podía implementar el modelo. 
- Pero la situación ha cambiado. Hoy hasta la computadora más simple puede ejecutar sofisticados sistemas de administración de bases de datos relacionales.
- La simplicidad de SQL, en donde un novato puede aprender a hacer consultas en poco tiempo. Es una gran parte de la razón de la popularidad del modelo relacional.

## Clase 13 Términos básicos de las bases de datos relacionales.
3. https://mariadb.com/kb/en/relational-databases-basic-terms/

El modelo de base de datos relacional usa ciertos términos para describir sus componentes:

- Los datos son los valores guardados en la base de datos. Por sí solos, los datos significan muy poco. CA 684-213 es un ejemplo de datos en una base de datos de la DMV (División de Vehículos Motorizados).
- La información es datos procesados. Por ejemplo, CA 684-213 es el número de registro del automóvil de un automóvil que pertenece a Lyndon Manson, en una base de datos de la DMV.
- Una base de datos es una colección de tablas , también llamadas entidades .
- Cada tabla está compuesta de registros (las filas horizontales de la tabla, también llamadas tuplas ). Cada registro debe ser único y puede almacenarse en cualquier orden en la tabla.
- Cada registro está formado por campos (que son las columnas verticales de la tabla, también llamadas atributos ). Básicamente, un registro es un hecho (por ejemplo, un cliente o una venta).
- Estos campos pueden ser de varios tipos . MariaDB tiene muchos tipos (consulte Tipos de datos para obtener una lista), pero generalmente los tipos se dividen en tres tipos: carácter, numérico y fecha. Por ejemplo, el nombre de un cliente es un campo de caracteres, el cumpleaños de un cliente es un campo de fecha y el número de hijos de un cliente es un campo numérico.
- El rango de valores permitidos para un campo se denomina dominio (también denominado especificación de campo ). Por ejemplo, un campo de tarjeta de crédito puede limitarse solo a los valores Mastercard, Visa y Amex.
- Se dice que un campo contiene un valor nulo cuando no contiene nada en absoluto. Los campos nulos pueden crear complejidades en los cálculos y tener consecuencias para la precisión de los datos. Por esta razón, muchos campos están configurados específicamente para que no contengan valores nulos.
- Una clave accede a registros específicos en una tabla.
- Un índice es un mecanismo para mejorar el rendimiento de una base de datos. Los índices a menudo se confunden con las claves. Los índices son, estrictamente hablando, parte de la estructura física, y las claves son parte de la estructura lógica. Sin embargo, a menudo verá los términos usados indistintamente, incluso en toda esta Base de conocimiento.
- Una vista es una tabla virtual compuesta por un subconjunto de las tablas reales.

- Una relación uno a uno (1: 1) es donde para cada instancia de la primera tabla en una relación, solo existe una instancia de la segunda tabla. Un ejemplo de esto sería un caso donde una cadena de tiendas lleva una venta máquina. Cada máquina expendedora solo puede estar en una tienda, y cada tienda tiene solo una máquina expendedora. one_to_one_relationship

- Una relación de uno a muchos (1: N) es donde para cada instancia de la primera tabla en una relación, existen muchas instancias de la segunda tabla. Este es un tipo común de relación. Un ejemplo es la relación entre un escultor y sus esculturas. Cada escultor puede haber creado muchas esculturas, pero cada escultura ha sido creada por un solo escultor. one_to_many_relationship

- Se produce una relación de muchos a muchos (M: N) donde, para cada instancia de la primera tabla, hay muchas instancias de la segunda tabla, y para cada instancia de la segunda tabla, hay muchas instancias de la primera. Por ejemplo, un estudiante puede tener muchos profesores, y un profesor puede tener muchos estudiantes. many_to_many_relationship

- Existe una relación obligatoria donde para cada instancia de la primera tabla en una relación, deben existir una o más instancias de la segunda . Por ejemplo, para que exista un grupo de música, debe existir al menos un músico en ese grupo.

- Una relación opcional es donde para cada instancia de la primera tabla en una relación, pueden existir instancias de la segunda. Por ejemplo, si un autor puede figurar en la base de datos sin haber escrito un libro (en otras palabras, un posible autor), esa relación es opcional. Sin embargo, lo contrario no es necesariamente cierto. Por ejemplo, para que un libro aparezca en la lista, debe tener un autor.

- La integridad de los datos se refiere a la condición en la que los datos son precisos, válidos y consistentes. Un ejemplo de integridad deficiente sería si el número de teléfono de un cliente se almacena de manera diferente en dos ubicaciones diferentes. Otro es donde un registro del curso contiene una referencia a un profesor que ya no está presente en la escuela. La normalización de la base de datos es una técnica que lo ayuda a minimizar el riesgo de este tipo de problemas.

### Clase 14. Zoom de términos básicos.
#### 1. Llaves https://mariadb.com/kb/en/relational-databases-table-keys/
- Una clave o índice , como lo indica el término en sí, desbloquea el acceso a las tablas. Si conoce la clave, sabe cómo identificar registros específicos y las relaciones entre las tablas.

- Cada clave consta de uno o más campos, o prefijo de campo. El orden de las columnas en un índice es significativo. Cada llave tiene un nombre.

- Una clave candidata es un campo, o combinación de campos, que identifica de forma exclusiva un registro. No puede contener un valor nulo y su valor debe ser único. (Con duplicados, ya no estaría identificando un registro único).

- Una clave primaria (PK) es una clave candidata que se ha designado para identificar registros únicos en la tabla en toda la estructura de la base de datos.

- Una clave sustituta es una clave primaria que contiene valores únicos generados automáticamente por el sistema de base de datos, generalmente, números enteros. Una clave sustituta no tiene significado, excepto la identificación exclusiva de un registro. Este es el tipo más común de clave primaria.

Por ejemplo, vea la siguiente tabla:

Código de cliente	Nombre de pila	Apellido	Número de teléfono
1	Ariane	Edison	448-2143
2	Georgei	Edison	448-2142
3	Abbas	Edison	9231-5312

- A primera vista, hay dos posibles claves candidatas para esta tabla. De cualquier customer_code o una combinación de nombre apellido , apellido y TELEPHONE_NUMBER serían suficientes. Siempre es mejor elegir la clave candidata con el menor número de campos para la clave primaria, por lo que elegiría customer_code en este ejemplo (tenga en cuenta que es una clave sustituta). Tras la reflexión, también existe la posibilidad de que la segunda combinación no sea única. La combinación de FIRST_NAME , apellido y TELEPHONE_NUMBERen teoría podría duplicarse, como cuando un padre tiene un hijo del mismo nombre que se puede contactar en el mismo número de teléfono. Este sistema tendría que excluir expresamente esta posibilidad para que estos tres campos sean considerados para el estado de la clave primaria.

- Puede haber muchas Edison Ariane, pero evita la confusión al asignar a cada una un número único. Una vez que se ha creado una clave principal, los candidatos restantes se etiquetan como claves alternativas .

La relación no es muy buena si en lugar de usar la clave primaria, customer_code , en la tabla de venta , usa otro campo que no es único, como el nombre del cliente. Es poco probable que sepa con certeza qué cliente realizó la venta en ese caso.

### Clase 15. Ciclo de vida de una base de datos.
https://mariadb.com/kb/en/database-lifecycle/

- Las bases de datos tienen una vida útil limitada.
- Nacen en una oleada de optimismo y avanzan por la vida logrando fama, fortuna y anonimato pacífico, o notoriedad según sea el caso, antes de desvanecerse una vez más. 
- Incluso la base de datos más exitosa en algún momento es reemplazada por otra estructura más flexible y actualizada, y así comienza la vida de nuevo. Aunque las definiciones exactas difieren, generalmente hay seis etapas del ciclo de vida de la base de datos.
- La fase de análisis es donde se entrevista a las partes interesadas y se examina cualquier sistema existente para identificar problemas, posibilidades y limitaciones. Se determinan los objetivos y el alcance del nuevo sistema.
- La fase de diseño es donde se crea un diseño conceptual a partir de los requisitos previamente determinados, y se crea un diseño lógico y físico que preparará la base de datos para la implementación.
- La fase de implementación es donde se instala el sistema de gestión de bases de datos (DBMS), se crean las bases de datos y se cargan o importan los datos.
- La fase de prueba es donde se prueba y ajusta la base de datos, generalmente junto con las aplicaciones asociadas.
- La fase de operación es donde la base de datos funciona normalmente, produciendo información para sus usuarios.
- La fase de mantenimiento es donde se realizan cambios en la base de datos en respuesta a nuevos requisitos o condiciones de operación cambiadas (como una carga más pesada).
- El desarrollo de bases de datos no es independiente del desarrollo de sistemas, a menudo es un componente del proceso de desarrollo de sistemas más amplio. Las etapas del desarrollo de sistemas básicamente reflejan las etapas del ciclo de vida de una base de datos, pero son un superconjunto. Mientras que el diseño de la base de datos se ocupa del diseño del sistema para almacenar los datos, el diseño de los sistemas también se ocupa de los procesos que afectarán los datos.

Según la información proporcionada, puede comenzar su diseño lógico y debería poder identificar las entidades iniciales:
•	Poeta
•	Poema
•	Publicación
•	Rebaja

The Poet's Circle no es una entidad, o incluso, por ejemplo, una entidad editorial . Solo si el sistema se desarrollara para muchos editores, el editor sería una entidad válida.

Ni el sitio web ni la comunidad de poesía son entidades. Solo hay un sitio web y, de todos modos, un sitio web es simplemente un medio de producir datos para poblar la base de datos. También hay una sola comunidad de poesía en lo que respecta a este sistema, y no hay mucho que quieras almacenar al respecto.

A continuación, debe determinar la relación entre estas entidades. Puede identificar lo siguiente:
- Un poeta puede escribir muchos poemas. El análisis identificó el hecho de que un poeta puede almacenarse en el sistema incluso si no hay poemas asociados. Los poemas pueden ser capturados en un momento posterior, o el poeta aún puede ser un poeta potencial. Por el contrario, muchos poetas podrían escribir un poema, aunque el poema debe haber sido escrito por al menos un poeta.

### Clase 20. Normalización una vista por encima
https://mariadb.com/kb/en/database-normalization-overview/

Desarrollado en la década de 1970 por EF Codd, la normalización de la base de datos es un requisito estándar de muchos diseños de bases de datos. La normalización es una técnica que puede ayudarlo a evitar anomalías en los datos y otros problemas con la administración de sus datos. Consiste en transformar una tabla a través de varias etapas: primera forma normal , segunda forma normal , tercera forma normal y más allá.

Apunta hacia:
- Elimine las redundancias de datos (y, por lo tanto, use menos espacio)
- Facilite los cambios en los datos y evite anomalías al hacerlo.
- Hacer que las restricciones de integridad referencial sean más fáciles de aplicar
- Produce una estructura fácilmente comprensible que se asemeja mucho a la situación que representan los datos y permite el crecimiento

Comencemos creando un conjunto de datos de muestra. Primero caminará por el proceso de normalización sin preocuparse por la teoría, para comprender las razones por las que desea normalizar. Una vez que haya hecho eso, presentaremos la teoría y las diversas etapas de normalización, lo que hará que todo el proceso descrito a continuación sea mucho más simple la próxima vez que lo haga.
