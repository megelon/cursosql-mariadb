Ahora que tenemos todo el diccionario por partes debemos almacenarlo en algún lado, ojalá seamos capaces de actualizarlo periódicamente. Las tablas son elementos vivos que cambian y se adaptan según las necesidades que surjan. Por ello tener un proceso automático que nos permita capturar los cambios nos ayudará a:
a.       Saber si hay información nueva que nos puede ser de utilidad. Con esto eliminamos la dependencia de una persona que nos tenga que informar cada vez que haga un cambio. Con un script  bastará. Esto nos ahorrará tiempo en reuniones y comunicaciones ineficientes.
b.       Auditar los cambios hechos a las tablas, si hay algo inconsistente lo podemos identificar ágilmente y podremos escalarlo con quien corresponda.
(Tener un responsable por tabla/esquema/base de datos es una buena práctica. Esto nos facilita el escalamiento de dudas/problemas y su pronta solución.)
En este punto es importante definir que es una inconsistencia:
-          Cambios del tipo en una variable. Ejemplo: el valor de las ventas ahora es un campo de tipo String y no del tipo entero. Esto es un problema porque podemos tener procesos automáticos que usan la variable de tipo entero. Un cambio de tipo implica agregar una función al script para transformar el tipo.
-          Cambios en el número de registros de la tabla. El número va a cambiar periódicamente pero generalmente seguirá patrones de comportamiento que podremos caracterizar con intervalos de confianza para saber que un cambio por fuera de las bandas del intervalo debe ser revisado. Ejemplo: el valor de las ventas puede verse afectado por eventos legítimos, pero también por eventos ilegítimos, tipo fraude o por algún error de configuración en donde los registros queden duplicados. Tener registros duplicados en una base de datos o tabla es algo que altera cualquier indicador. Luego veremos cómo evitar que dupliquemos registros y trabajemos con ellos erradamente.
