Lo único que necesitas para aprender SQL es practicar. Sin embargo, aprender por aprender no resulta eficiente. Es más fácil cuándo existe una necesidad y adquieres un conocimiento específico para satisfacerla.

Bajo este contexto, vamos a hablar sobre una metodología que prioriza la identificación de una necesidad de negocio y nos facilitará el proceso de aprendizaje. Esta metodología, ha sido justamente pensada para facilitar el trabajo en proyectos que hacen uso intensivo de datos.
Su nombre es CRISP-DM por sus siglas en ingles Cross Industry Standard Process for Data Mining. Está vigente desde 1999 y continúa siendo la metodología más usada para proyectos que involucran uso intensivo de datos. 
Esta metodología captura las fases claves de un proceso en el que se realiza uso de datos para atender una necesidad de negocio. Sus fases comprenden: 
- entendimiento de los datos, 
- preparación de los datos, 
- modelaje, 
- evaluación,
- y finalmente producción, 
resaltando el proceso iterativo entre fases, tal como lo podemos ver en la ilustración.

Clase 4. Detalle de la metodología CRISP DM
Si tenemos un problema de reportería y una solución es montar un tablero, no es suficiente con tener un buen diseño, montarlo sobre la mejor herramienta y dar acceso a los empleados, si al final, nadie tiene idea de cómo interactuar con él para sacarle provecho. Es importante garantizar la adopción de la solución hasta volver expertos a unos cuantos, que luego vuelvan expertos a otro puñado; y poco a poco habilitar capacidades en la compañía que permitan ver el cambio y aceleren el retorno de la inversión.
Otra arista clave en cualquier implementación es hacerla viable. Para eso te recomiendo dividir el proceso en etapas, tal como lo sugieren las metodologías ágiles, que invitan a ir sacando a producción productos/soluciones de forma iterativa e incremental. Por ejemplo, si tu objetivo es construir un carro, la mejor forma de abordar el proyecto será dividiéndolo en fases. Puedes empezar con un producto mínimo viable e ir agregando valor con cada nueva entrega. La definición de un producto mínimo viable es un producto útil que no es el entregable final, pero agrega valor a los usuarios en el corto plazo. Imagína que dispones de un mes para la construcción del carro, entonces divides el proyecto en 4 sprints o entregables.
- Sprint 1 (primera semana): El objetivo de este sprint es armar la carrocería básica del carro. Ponerle ruedas, timón y una silla en el puesto del conductor. En este momento tendrás una versión muy incipiente del carro, pero te permitirá probar el tamaño, la distribución del espacio, la resistencia de los materiales, entre otras cosas. 
- Sprint 2 (segunda semana): Ahora el objetivo es construir el motor del carro y montarlo sobre la carrocería de la semana pasada. Con esto tendrás la primera versión funcional del carro. Podrás encenderlo, probar el motor y ponerlo a rodar. No tendrás que esperar hasta el final del mes para validar si el carro funciona. Desde la semana dos, podrás detectar aspectos a mejorar.
- Sprint 3 (tercera semana): esta semana puedes instalar las sillas del resto de pasajeros, ponerle puertas y ventanas al carro, instalar los cinturones de seguridad, entre otros detalles. Con esto podrás sacarlo a la calle y hacer un viaje completo. Puede que te aburras sin poder poner música o aire acondicionado, pero esto podrás solucionarlo en la siguiente entrega.
- Sprint 4 (cuarta semana): Ya que tienes un carro funcional, puedes dedicar esta semana para hacerlo un poco más lujoso. Instalarle un radio y parlantes, ponerle cojinería de cuero a las sillas, pintarlo del color que prefieras y agregarle detalles que lo hagan atractivo.
Cada entrega agrega valor, ya que en cada una tendrás nuevas funcionalidades que te permitirán probar y fallar rápido y obtendrás lecciones aprendidas que retroalimentarán la siguiente etapa; y el cliente final tendrá disponible un producto útil desde fases tempranas del proyecto. # Analogía empresa de abogados
Para entender las herramientas que vamos a usar en este curso veamos la siguiente analogía:
La complejidad de gestión documental en una empresa de abogados es alta y puede resultar desastrosa. Para evitar esto, debemos hacer algo que permita: 
- Reducir la complejidad en la gestión 
- Consultar cualquier documento de forma rápida
- Y saber rápidamente dónde queda almacenado cada documento.
Todos los abogados compartimos el mismo archivo y aunque algunos sean ordenados, no todos tienen la buena práctica. El día a día nos consume. Todos guardamos y consultamos documentos constantemente. Eventualmente nuestro orden se verá afectado y puede que perdamos nuestros archivos.
Está claro que necesitamos de alguien que nos ayude a guardarlos de forma ordenada y a mantener el orden en el tiempo. Para ello contratamos un experto, diseñamos una biblioteca para almacenamiento de carpetas y una ventanilla, ya que restringimos el acceso. Únicamente el experto tiene acceso a los documentos almacenados.
Como la empresa tiene tantos abogados y maneja tantos documentos, el experto debe ser ágil, capaz de entregar documentos en el menor tiempo posible ya que no hay tiempo que perder. Él facilitará el trabajo de cada abogado, entregará justo el documento solicitado, e incluso si se lo piden, podrá entregar un set de documentos. Siempre en un tiempo asombroso.
La única limitante es que el experto habla otro idioma y debemos aprenderlo para comunicarnos con él. El idioma usa convenciones que le permiten entender: 
a) dónde está cada cosa 
b) y que tiene que hacer para entregarnos justo lo que queremos.
En esta analogía: 
-	El experto, es nuestro motor de base de datos o sistema de gestión de bases de datos relacional. 
-	La ventanilla por donde dejamos y recibimos los pedidos es nuestro cliente de base datos.
-	El idioma del experto es SQL (Structured Query Language).

#### MariaDB
